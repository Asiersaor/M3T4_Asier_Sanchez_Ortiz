## 1. Clase Pedido con objeto dentro(Composicion basica)

```python
class cliente:
    def __init__(self):
        self._nombre = None
        self._direccion = None
        self._telefono = None
    @property
    def nombre(self):
        return self._nombre
    @property
    def direccion(self):
        return self._direccion
    @property
    def telefono(self):
        return self._telefono
    @nombre.setter
    def nombre(self, valor):
        self._nombre = valor
    @direccion.setter
    def direccion(self, valor):
        self._direccion = valor
    @telefono.setter
    def telefono(self, valor):
        self._telefono = valor
class pedido:
    def __init__(self):
        self._codigo = None
        self._producto = None
        self._cantidad = None
        self._precio_unitario = None 
        self._cliente = None
    @property
    def codigo(self):
        return self._codigo
    @property
    def producto(self):
        return self._producto
    @property
    def cantidad(self):
        return self._cantidad
    @property
    def precio_unitario(self):
        return self._precio_unitario
    @property
    def cliente(self):
        return self._cliente
    @codigo.setter
    def codigo(self, valor):
        self._codigo = valor
    @producto.setter
    def producto(self, valor):
        self._producto = valor
    @cantidad.setter
    def cantidad(self, valor):
        self._cantidad = valor
    @precio_unitario.setter
    def precio_unitario(self, valor):
        self._precio_unitario = valor
    @cliente.setter
    def cliente(self, valor):
        self._cliente = valor
    def calcular_total(self):
        return f"El precio total del pedido es de {self._cantidad * self._precio_unitario} €"
    def __str__(self):
        return f"Datos de cliente:\nNombre: {self._cliente.nombre}\nDireccion: {self._cliente.direccion}\nTeleno: {self._cliente.telefono}\nDatos del pedido:\nCodigo: {self._codigo}\nProducto: {self._producto}\nCantidad: {self._cantidad}\nPrecio unitario: {self._precio_unitario} €"
def main():
    mi_cliente = cliente()
    mi_pedido = pedido()
    mi_pedido.cliente = mi_cliente
    mi_cliente.nombre = "Asier Sanchez"
    mi_cliente.direccion = "C/ Sahagun Madrid"
    mi_cliente.telefono = 640257938
    mi_pedido.codigo = "#1"
    mi_pedido.producto = "Bateria portatil"
    mi_pedido.cantidad = 4
    mi_pedido.precio_unitario = 12.99
    print(mi_pedido)
    print(mi_pedido.calcular_total())
    mi_pedido.codigo = "#7"
    mi_pedido.producto = "Cable tipo C"
    mi_pedido.precio_unitario = 4.55
    mi_cliente.nombre = "Jose Antonio"
    print(mi_pedido)
    print(mi_pedido.calcular_total())
main()
```

## 2. Ampliación: Pedido con múltiples productos y cantidades (composición avanzada).

```python
class cliente:
    def __init__(self):
        self._nombre = None
        self._direccion = None
        self._telefono = None
    @property
    def nombre(self):
        return self._nombre
    @property
    def direccion(self):
        return self._direccion
    @property
    def telefono(self):
        return self._telefono
    @nombre.setter
    def nombre(self, valor):
        self._nombre = valor
    @direccion.setter
    def direccion(self, valor):
        self._direccion = valor
    @telefono.setter
    def telefono(self, valor):
        self._telefono = valor
class producto:
    def __init__(self):
        self._codigo = None
        self._nombre = None
        self._precio_unitario = None
    @property
    def codigo(self):
        return self._codigo
    @property
    def nombre(self):
        return self._nombre
    @property
    def precio_unitario(self):
        return self._precio_unitario
    @codigo.setter
    def codigo(self, valor):
        self._codigo = valor
    @nombre.setter
    def nombre(self, valor):
        self._nombre = valor
    @precio_unitario.setter
    def precio_unitario(self, valor):
        self._precio_unitario = valor
class pedido:
    def __init__(self):
        self._codigo = None
        self._producto = []
        self._cantidad = []
        self._cliente = None
    @property
    def codigo(self):
        return self._codigo
    @property
    def producto(self):
        return self._producto
    @property
    def cantidad(self):
        return self._cantidad
    @property
    def cliente(self):
        return self._cliente
    @codigo.setter
    def codigo(self, valor):
        self._codigo = valor
    @producto.setter
    def producto(self, valor):
        self._producto = valor
    @cantidad.setter
    def cantidad(self, valor):
        self._cantidad = valor
    @cliente.setter
    def cliente(self, valor):
        self._cliente = valor
    def agregar_producto(self, producto, cantidad):
        self._producto.append(producto)
        self._cantidad.append(cantidad)
    def calcular_total(self):
        valor = 0
        for i in range(len(self._cantidad)):
            valor = valor + self._producto[i].precio_unitario * self._cantidad[i]
        return f"El precio total del pedido es de {valor} €"
    def mostrar_pedido(self):
        for i in range(len(self._producto)):
            print(f"Nombre de producto: {self._producto[i].nombre} | Cantidad: {self._cantidad[i]} | Precio unitario: {self._producto[i].precio_unitario} | Codigo de producto: {self._producto[i].codigo}")
    def __str__(self):
        return f"Datos de cliente:\nNombre: {self._cliente.nombre}\nDireccion: {self._cliente.direccion}\nTeleno: {self._cliente.telefono}\nDatos del pedido:\nCodigo: {self._codigo}\nProducto: {len(self.producto)}\nCantidad: {self._cantidad}"
def main():
    mi_cliente = cliente()
    mi_producto = producto()
    mi_pedido = pedido()
    mi_pedido.cliente = mi_cliente
    mi_cliente.nombre = "Asier Sanchez"
    mi_cliente.direccion = "C/ Sahagun Madrid"
    mi_cliente.telefono = 640257938
    mi_pedido.codigo = "#145"
    mi_producto.codigo = "#1"
    mi_producto.nombre = "Bateria Portatil"
    mi_producto.precio_unitario = 12.99
    mi_pedido.agregar_producto(mi_producto, 4)
    mi_producto2 = producto()
    mi_producto2.codigo = "#7"
    mi_producto2.nombre = "Cable tipo C"
    mi_producto2.precio_unitario = 4.55
    mi_pedido.agregar_producto(mi_producto2, 2)
    print(mi_pedido)
    print(mi_pedido.calcular_total())
    mi_pedido.mostrar_pedido()
main()
```

## 3. Herencia simple en un sistema de viviendas

```python
class Vivienda():
    def __init__(self):
        self._direccion = None
        self._metros_cuadrados = None
        self._precio = None
    @property
    def direccion(self):
        return self._direccion
    @property
    def metros_cuadrados(self):
        return self._metros_cuadrados
    @property
    def precio(self):
        return self._precio
    @direccion.setter
    def direccion(self, valor):
        self._direccion = valor
    @metros_cuadrados.setter
    def metros_cuadrados(self, valor):
        self._metros_cuadrados = valor
    @precio.setter
    def precio(self, valor):
        self._precio = valor
    def __str__(self):
        return f"Direccion: {self._direccion} | Metros cuadrados: {self._metros_cuadrados}m2 | Precio: {self._precio}€"
class Casa(Vivienda):
    def __init__(self):
        super().__init__()
        self._jardin = None 
    @property
    def jardin(self):
        return self._jardin
    @jardin.setter
    def jardin(self, valor):
        if valor:
            valor = "Si"
        else:
            valor = "No"
        self._jardin = valor
    def __str__(self):
        return f"{super().__str__()} | Jardin: {self._jardin}"
class Apartamento(Vivienda):
    def __init__(self):
        super().__init__()
        self._piso = None
    @property
    def piso(self):
        return self._piso
    @piso.setter
    def piso(self, valor):
        self._piso = valor
    def __str__(self):
        return f"{super().__str__()} | Piso: Numero {self._piso}"
def main():
    mi_Casa = Casa()
    mi_Apartamento = Apartamento()
    mi_Casa.direccion = "C/ Francos Rodriguez Portal 4"
    mi_Casa.metros_cuadrados = 64.5
    mi_Casa.precio = 325000
    mi_Casa.jardin = False
    mi_Apartamento.direccion = "C/ Del Pou 15"
    mi_Apartamento.metros_cuadrados = 79.43
    mi_Apartamento.precio = 32000
    mi_Apartamento.piso = "3"
    print(mi_Casa)
    print(mi_Apartamento)
main()
```
